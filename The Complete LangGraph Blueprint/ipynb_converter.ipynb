```python
import os
import py2nb  # Ensure this module exists in your environment

root_dir = "The Complete LangGraph Blueprint"

for folder, _, files in os.walk('.'):
    for file in files:
        if file.endswith(".py") and file != "display_graph.py":
            relative_path = os.path.join(folder, file)
            try:
                notebook_path = py2nb.convert(relative_path)
                print(f"Converted: {relative_path} -> {notebook_path}")
            except Exception as e:
                print(f"Error converting {relative_path}: {e}")
```
